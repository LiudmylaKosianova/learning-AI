Regression model with Keras project

I start with importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import keras
from keras.models import Sequential
from keras.layers import Dense
```


Task 1. Randomly split the data into a training and test set by holding 30% of the data for testing.


```python
#First I download the data and read it into a pandas dataframe and look at the file
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
#concrete_data.head()

#Then I split the data into predictors (x) and target (y)
x = concrete_data.drop('Strength', axis=1)
y = concrete_data['Strength']

#x.head()
#y.head()

#Then I normalize the data, substracting the mean and dividing by the standart deviation
xNorm = (x - x.mean())/x.std()
#xNorm.head()

#Then I save the number of predictors
n_cols = xNorm.shape[1]


#Now I can split the data into test and train set
#I use train_test_split() function from scikit-learn
xNorm_train, xNorm_test, y_train, y_test = train_test_split(xNorm, y, test_size=0.3, random_state=42)
```


Task 2. Train the model on the training data using 50 epochs


```python
#First I will create a model
model = Sequential()

#Then I use add method to add a hidden layer with 10 nodes
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))

#Then I use add method to add output layer
model.add(Dense(1))

#Then I define optimiser for training and error metric
model.compile(optimizer='adam', loss='mean_squared_error')

#Then I train the model on the training data using 50 epochs
model.fit(xNorm_train, y_train, epochs=50)
```


Task 3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete stregth and the actual concrete strength


```python
#First I will evaluate the model with predict method
predictions = model.predict(xNorm_test)

#Then I will compute the mean squared error
msEr = mean_squared_error(y_test, predictions)

#print(msEr)
```


Task 4. Repeat tasks 1-3 50 times, creating a list of 50 mean squared errors.


```python
#I will create an empty list for mean squared errord
msEr50 = []

#I will create a for loop that will iterate 50 times to repeat tasks 1-3
#I will append new value of mean squared errors to my list msEr50

for i in range(50):
#randomly split the data
    xNorm_train, xNorm_test, y_train, y_test = train_test_split(xNorm, y, test_size=0.3, random_state=42)
    
#create a model
    model = Sequential()

#add a hidden layer with 10 nodes
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))

#add two more hidden layers with 10 nodes each
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))

#add output layer
    model.add(Dense(1))

#define optimiser for training and error metric
    model.compile(optimizer='adam', loss='mean_squared_error')

#train the model on the training data using 50 epochs
    model.fit(xNorm_train, y_train, epochs=100)
    
#test the model
    predictions = model.predict(xNorm_test)
    
#compute the mean squared error
    msEr = mean_squared_error(y_test, predictions)
#add the value to the list
    msEr50.append(msEr)    
```


Task 5. Report the the mean and the standard deviation of the mean squared errors


```python
#I will convert my list msEr50 into a NumPy array
mse50_array = np.array(msEr50)

#I will calculate the mean of mean squared errors
mean_mse = np.mean(mse50_array)

#I will calculate the standard deviation of the mean squared errors
std_mse = np.std(mse50_array)

print(f"Mean MSE: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
```
